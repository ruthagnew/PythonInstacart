# 3. Merging orders dataset with prior orders dataset

We are going to merge our clean dataset, orders, with a new dataset called *prior orders* which will give us information on the product id. Our join will be made on the shared variable, *order_id*. 

```python
#merging df_ords and df_ords_prior
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
#checking output
df_merged_large.head()
```

```python
#checking output
df_merged_large.shape
```

```python
df_merged_large['_merge'].value_counts()
```

```python
df_merged_large.describe()
```

**Observations**

Our new dataframe has the same number of rows as before, so the merge is successful. 

# 4. Combine newly merged dataframe with products dataset

Merging with the products dataset will give us crucial information about the products purchased in each order. Using the merge() function on the common variable, 'product_id' will enable us to see any instances where the product_id doesn't match, as they would appear as NaN values. 

```python
#merging data sets
df_merged_final = df_merged_large.merge(df_prods, on = ['product_id'])
```

```python
df_merged_final.head()
```

```python
#check output
df_merged_final['_merge'].value_counts()
```

**Observations**

The dataset has the same number of rows, so the merge was successful.
