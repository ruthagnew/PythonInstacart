# 3. Importing departments dataset

We need the departments dataset so we can cross reference the department id with department name. This will help later in our analysis when we want to know which departments account for the most spends. 

### A. Importing dataset

```python
#importing data set departments.csv
df_dep = pd.read_csv(os.path.join(path, '02_Data', 'Original_data', 'departments.csv'), index_col = False)
```

### B. Transposing dataset

```python
#checking output of dataset
df_dep.head()
```

```python
#transposing dataset for future merging
df_dep.T
```

```python
#saving newly transposed dataset as df_dept_t
df_dep_t = df_dep.T
```

```python
#checking output 
df_dep_t
```

```python
#adding an index
df_dep_t.reset_index()
```

### C. Changing the header

When we transposed the data, the header row became row 0. We now need to reset the header. 

```python
#take the first row of the df_dep_t for the header
new_header = df_dep_t.iloc[0]
```

```python
#checking output
new_header
```

```python
#take the data under the header row for a new df
df_dep_t_new = df_dep_t[1:]
```

```python
#return the new df
df_dep_t_new
```

```python
#set the header row as the df header
df_dep_t_new.columns = new_header
```

```python
#checking output
df_dep_t_new
```

### D. Data dictionaries

We need to create a data dictionary of the department ids and department names so we can use this in our future analysis. 

```python
#creating data dictionary
data_dict = df_dep_t_new.to_dict('index')
```

```python
#checking output
data_dict
```

```python
#using the data dictionary to understand values in the data frame
df_prods.head()
```

```python
#checking output of one department
print(data_dict.get('19'))
```

### E. Testing out subsetting on new data dictionary 

Testing out creating subsets of data dictionary using loc and isin functions. 

```python
#using the loc function to subsetting
df_snacks_2 = df_prods.loc[df_prods['department_id']==19]
```

```python
df_snacks_2.head()
```

```python
#using loc combined with isin to search multiple values
df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([17, 18, 19])]
```

```python
df_snacks_3
```
